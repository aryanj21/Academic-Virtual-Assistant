```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

pdf_path = "pdf_files/SMA.pdf"  # Update this with the path to your PDF
raw_text = extract_text_from_pdf(pdf_path)
print(raw_text)  # To see the extracted text
```

```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt')

def clean_text(text):
    # Remove unwanted characters like newlines, tabs, etc.
    cleaned_text = text.replace('\n', ' ').replace('\t', ' ').strip()
    return cleaned_text

def chunk_text(cleaned_text, chunk_size=5):
    sentences = sent_tokenize(cleaned_text)
    chunks = [' '.join(sentences[i:i+chunk_size]) for i in range(0, len(sentences), chunk_size)]
    return chunks

cleaned_text = clean_text(raw_text)
text_chunks = chunk_text(cleaned_text)
print(text_chunks)  # To see the processed chunks
```

```python
from sentence_transformers import SentenceTransformer

def generate_embeddings(chunks):
    model = SentenceTransformer('all-MiniLM-L6-v2')  # This is the pre-trained model for embedding
    embeddings = model.encode(chunks)
    return embeddings

embeddings = generate_embeddings(text_chunks)
print(embeddings.shape)  # To check the shape of embeddings
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_most_similar(query, embeddings, chunks):
    model = SentenceTransformer('all-MiniLM-L6-v2')  # Use the same model as before
    query_embedding = model.encode([query])
    similarities = cosine_similarity(query_embedding, embeddings)
    most_similar_idx = np.argmax(similarities)
    return chunks[most_similar_idx]

query = "Explain the process of photosynthesis"  # You can change the query as per your test case
response = get_most_similar(query, embeddings, text_chunks)
print("Response:", response)
```

```python
class ChatbotMemory:
    def __init__(self):
        self.history = []
    
    def add_to_memory(self, user_input, response):
        self.history.append((user_input, response))
    
    def get_memory(self):
        return self.history
    
    def print_memory(self):
        for user_input, response in self.history:
            print(f"User: {user_input}\nBot: {response}\n")

chat_memory = ChatbotMemory()

def chat(query):
    response = get_most_similar(query, embeddings, text_chunks)
    chat_memory.add_to_memory(query, response)
    return response

# Simulating a conversation in Jupyter Notebook
user_input = "Explain how photosynthesis works."
bot_response = chat(user_input)
print(f"Bot: {bot_response}")

# Display the memory
chat_memory.print_memory()
```

```python
import gradio as gr

def chatbot_interface(user_input):
    return chat(user_input)

interface = gr.Interface(fn=chatbot_interface, inputs="text", outputs="text", title="Academic Chatbot")
interface.launch()
```
